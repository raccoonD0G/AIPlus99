- [X] AG_News dataset 준비
	- Huggingface dataset의 `fancyzhx/ag_news`를 load
	- `collate_fn` 함수에 다음 수정사항들을 반영
    - Truncation과 관련된 부분들을 삭제
- [X] Classifier output, loss function, accuracy function 변경
	- 뉴스 기사 분류 문제는 binary classification이 아닌 일반적인 classification 문제입니다. MNIST 과제에서 했던 것 처럼 `nn.CrossEntropyLoss` 를 추가하고 `TextClassifier`의 출력 차원을 잘 조정하여 task를 풀 수 있도록 수정
	- 그리고 정확도를 재는 `accuracy` 함수도 classification에 맞춰 수정
- [X]  학습 결과 report
    - DistilBERT 실습과 같이 매 epoch 마다의 train loss를 출력하고 최종 모델의 test accuracy를 report 첨부

```python
import torch
from datasets import load_dataset
from torch.utils.data import DataLoader
from torch.optim import Adam
import matplotlib.pyplot as plt
from torch import nn
from tqdm import tqdm

def accuracy(model, dataloader):
    cnt = 0
    acc = 0

    model.eval()

    for input_ids, attention_mask, labels in dataloader:
        input_ids = input_ids.to('cuda')
        attention_mask = attention_mask.to('cuda')
        labels = labels.to('cuda')

        preds = model(input_ids, attention_mask)
        preds = torch.argmax(preds, dim=-1)

        cnt += labels.shape[0]
        acc += (labels == preds).sum().item()

    return acc / cnt


# 데이터를 배치로 묶기 위한 함수 정의
def collate_fn(batch):
    texts, labels = [], []

    for row in batch:
        texts.append(row['text'])
        labels.append(row['label'])

    # truncation 제거, max_length 제거
    encoded = tokenizer(texts, padding=True, return_tensors="pt")

    input_ids = encoded["input_ids"]
    attention_mask = encoded["attention_mask"]
    labels = torch.LongTensor(labels)

    return input_ids, attention_mask, labels

# 텍스트 분류 모델 정의 (DistilBERT + Linear layer)
class TextClassifier(nn.Module):
    def __init__(self):
        super().__init__()

        # 사전학습된 DistilBERT 모델을 encoder로 불러옴 (pretrained transformer)
        self.encoder = torch.hub.load('huggingface/pytorch-transformers', 'model', 'distilbert-base-uncased')

        # [CLS] 토큰 분류기 정의
        self.classifier = nn.Sequential(
            nn.Linear(768, 512),
            nn.LayerNorm(512),
            nn.LeakyReLU(0.1),
            nn.Dropout(0.3),

            nn.Linear(512, 256),
            nn.LayerNorm(256),
            nn.LeakyReLU(0.1),
            nn.Dropout(0.3),

            nn.Linear(256, 128),
            nn.LayerNorm(128),
            nn.LeakyReLU(0.1),
            nn.Dropout(0.3),

            nn.Linear(128, 4),
        )

    def forward(self, input_ids, attention_mask):
        x = self.encoder(input_ids=input_ids, attention_mask=attention_mask)['last_hidden_state']
        x = self.classifier(x[:, 0])
        return x

# DistilBERT 모델용 tokenizer 로드 (pretrained)
# 이 tokenizer는 문장을 토큰화해서 모델이 이해할 수 있는 input_ids로 변환해줌
tokenizer = torch.hub.load('huggingface/pytorch-transformers', 'tokenizer', 'distilbert-base-uncased')

# IMDB 감정 분석 데이터셋의 5%만 로드 (학습 데이터와 테스트 데이터 각각)
train_ds = load_dataset("fancyzhx/ag_news", split="train[:5%]")
test_ds = load_dataset("fancyzhx/ag_news", split="test[:5%]")

# 학습용 DataLoader 정의 (shuffle=True로 배치 순서 랜덤화)
train_loader = DataLoader(
    train_ds, batch_size=128, shuffle=True, collate_fn=collate_fn
)

# 테스트용 DataLoader 정의 (shuffle=False로 배치 순서 고정)
test_loader = DataLoader(
    test_ds, batch_size=128, shuffle=False, collate_fn=collate_fn
)

#DistilBERT 모델을 PyTorch Hub에서 로드 후 model 출력
bert = torch.hub.load('huggingface/pytorch-transformers', 'model', 'distilbert-base-uncased')

model = TextClassifier()

for name, param in model.encoder.named_parameters():
    if "transformer.layer.4" in name or "transformer.layer.5" in name:
        param.requires_grad = True
    else:
        param.requires_grad = False

# 정확도 및 손실 기록용 리스트
train_losses = []
train_accuracies = []
test_accuracies = []

# 학습 설정
lr = 0.001
model = model.to('cuda')  # 모델을 GPU로 이동
loss_fn = nn.CrossEntropyLoss()

optimizer = Adam(model.parameters(), lr=lr)
n_epochs = 50

# 학습 루프
for epoch in range(n_epochs):
    total_loss = 0.
    model.train()  # 학습 모드 설정

    for input_ids, attention_mask, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{n_epochs}"):
        model.zero_grad()  # 이전 gradient 초기화
        input_ids, attention_mask, labels = input_ids.to("cuda"), attention_mask.to("cuda"), labels.to("cuda")
        preds = model(input_ids, attention_mask)
        loss = loss_fn(preds, labels)
        loss.backward()  # 역전파
        optimizer.step()  # 파라미터 업데이트

        total_loss += loss.item()  # loss 누적

    train_losses.append(total_loss)
    tqdm.write(f"Epoch {epoch:3d} | Train Loss: {total_loss}")

    # 평가 시 gradient 계산 비활성화
    with torch.no_grad():
        model.eval()  # 평가 모드로 전환 (계산 비활성화)
        train_acc = accuracy(model, train_loader)
        test_acc = accuracy(model, test_loader)

        train_accuracies.append(train_acc)
        test_accuracies.append(test_acc)
        tqdm.write(f"=========> Train acc: {train_acc:.3f} | Test acc: {test_acc:.3f}")


plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.plot(train_losses, label="Train Loss")
plt.title("Train Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid()
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(train_accuracies, label="Train Acc")
plt.plot(test_accuracies, label="Test Acc")
plt.title("Accuracy Curve")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

# **Report**

## CrossEntropy

---

> 분류 문제에서 주로 사용되는 Loss 함수입니다. Softmax 함수와 세트를 이룹니다. 때문에 현재 모델의 마지막 레이어에는 암시적으로 Softmax 가 적용되었다고 볼 수 있습다. Softmax 는 출력값들의 총 합을 1로 만들어 분류 문제에서 확률 분포를 형성하는 용도로 쓰입니다.

> 반면 실습 파일에 있던 BCEWithLogitsLoss 는 Sigmoid 와 세트를 이룹니다. Sigmoid 는 0에서 1 사이의 값을 가져 확률과 비슷한 인상을 줍니다. 따라서 이진 분류 문제에서 확률을 구하는 용도로 사용하게 됩니다.

> 과제 목표는 4개중 하나로 분류하는 것으로, Softmax 를 포함한 CrossEntropy 가 적합합니다.

## Truncation과 관련된 부분들을 삭제

---

> collate_fn 함수에서 관련한 부분을 삭제했습다. Truncation 는 문장을 일정 길이로 잘라 안정적 학습을 돕습는다. 하지만 잘려나가는 만큼 정보 손실이 있을 수 있습다. DistilBERT 는 내부적으로 최대 입력 길이(512 토큰)를 알아서 관리하기 때문에, 굳이 Truncation 를 할 필요가 없습니다.

## 학습 결과

---

> Train acc 와 Test acc 가 모두 90% 이상을 기록하는 모습입니다. Dropout과 LayerNorm의 도움으로 과적합은 적은 것으로 보입니다. 전체적으로 안정적인 학습을 보여주나, 도중에 정확도가 소폭 감소하는 것으로 보아 Weight Decay등의 다른 정규화 방법을 고려할 수 있습니다.
